# Квантиль:

```python
df['danceability'].quantile([0.005,.01,.05,.1,.5,.9,.95,.99,.995])
```

Удалим строки, в которых danceability выходит за пределы квантилей 0.05 и 0.95

```python
rows_to_drop = df[(df['danceability'] < df['danceability'].quantile(0.05)) | (df['danceability'] > df['danceability'].quantile(0.95))].index
df = df.drop(rows_to_drop)

df
```

# Обработка данных

Здесь мне нужно удалить столбцы artist и song, а также заменить значения explicit на 0 и 1. После этого приводим все данные к типу float

```python
import pandas as pd

df_cleaned = df.drop(columns=['artist', 'song', 'genre'])

df_cleaned['explicit'] = df_cleaned['explicit'].replace({'True': 1, 'False': 0})

df_cleaned = df_cleaned.astype(float)

df_cleaned
```

# Нормализуем данные

```python
from sklearn.preprocessing import MinMaxScaler

scaler = MinMaxScaler()

df_normalized = pd.DataFrame(scaler.fit_transform(df_cleaned), columns=df_cleaned.columns)

df_normalized
```

# Корреляционная матрица:

```python
import seaborn as sns
import matplotlib.pyplot as plt

correlation_matrix = df_normalized.corr()

plt.figure(figsize=(12, 10))

sns.heatmap(correlation_matrix, square=True, cmap="coolwarm", fmt=".2f", linewidths=0.5)
plt.title("Correlation Matrix")
plt.show()
```

```python
from sklearn.model_selection import train_test_split
from sklearn.neighbors import KNeighborsRegressor
from sklearn.metrics import mean_squared_error, r2_score

x = df_normalized.drop(columns=['danceability'])
y = df_normalized['danceability']

x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.3, random_state=42)

knn = KNeighborsRegressor(n_neighbors=4)

knn.fit(x_train, y_train)

y_pred = knn.predict(x_test)

mse = mean_squared_error(y_test, y_pred)
r2 = r2_score(y_test, y_pred)

print("Mean Squared Error:", mse)
print("R^2 Score:", r2)
```

# Вывод:


*   **Mean Squared Error**: Среднеквадратичная ошибка равна 0.04997, что показывает среднее квадратичное отклонение предсказанных значений от фактических.
*   **R² Score**: Коэффициент детерминации составляет 0.1430, что указывает на невысокую точность модели.

Исходя из этого, можно сказать, что модель имеет довольно низкую точность



# КЛАСТЕРИЗАЦИЯ (Дз для 23.12)

Кластеризация KMeans

```python
from sklearn.model_selection import GridSearchCV
from sklearn.cluster import KMeans
from sklearn.metrics import silhouette_score, silhouette_samples
from sklearn.metrics import make_scorer

import warnings
warnings.simplefilter(action='ignore', category=FutureWarning)
warnings.simplefilter(action='ignore', category=UserWarning)

parameters = {'n_clusters': list(range(80, 200))}
km = KMeans()
clf = GridSearchCV(km, parameters)
clf.fit(x)
clusters_kmeans = clf.best_params_['n_clusters']
clf.best_params_
```

```python
kmeans = KMeans(n_clusters=clusters_kmeans, random_state=42)
df_normalized['KMeans_Cluster'] = kmeans.fit_predict(x)
silhouette_kmeans = silhouette_score(x, df_normalized['KMeans_Cluster'])
print(f"Silhouette score для KMeans: {silhouette_kmeans}")
```

```python
sns.pairplot(df_normalized, vars=["energy", "loudness"], hue='KMeans_Cluster', palette='magma')
plt.show()
```

```python
import numpy as np

labels = kmeans.labels_
silhouette_vals = silhouette_samples(x, labels)

n_clusters = len(set(labels))
silhouette_means = [np.mean(silhouette_vals[labels == i]) for i in range(n_clusters)]

plt.figure(figsize=(8, 5))
plt.bar(range(n_clusters), silhouette_means, color='green')
plt.xlabel('Кластеры')
plt.ylabel('Средний Silhouette score')
plt.title('Средний Silhouette score по кластерам KMeans')
plt.show()
```

Кластеризация DBScan

```python
from sklearn.cluster import DBSCAN
def silhouette_scorer(estimator, X):
    labels = estimator.fit_predict(x)
    if len(set(labels)) > 1:
        return silhouette_score(x, labels)
    else:
        return -1
parameters = {
    'eps': [1.0, 2.5, 3.0],
    'min_samples': [15, 20, 25]
}
dbscan = DBSCAN()
clf_dbscan = GridSearchCV(dbscan, parameters, scoring=make_scorer(silhouette_scorer))
clf_dbscan.fit(x)
eps_dbscan = clf_dbscan.best_params_['eps']
min_samples_dbscan = clf_dbscan.best_params_['min_samples']
clf_dbscan.best_params_
```

```python
dbscan = DBSCAN(eps=eps_dbscan, min_samples=min_samples_dbscan)
df_normalized['DBSCAN_Cluster'] = dbscan.fit_predict(x)
silhouette_dbscan = silhouette_score(x, df_normalized['DBSCAN_Cluster'])
print(f"Silhouette score для DBSCAN: {silhouette_dbscan}")
```

```python
sns.pairplot(df_normalized, vars=["energy", "loudness"], hue='DBSCAN_Cluster', palette='magma')
plt.show()
```

Кластеризация Agglomerative Clustering

```python
from sklearn.cluster import AgglomerativeClustering

parameters = {
    'n_clusters': list(range(100, 120))
}

agg_clustering = AgglomerativeClustering()

clf_agg = GridSearchCV(agg_clustering, parameters, scoring=make_scorer(silhouette_scorer))
clf_agg.fit(x)
clusters_agg = clf_agg.best_params_['n_clusters']
clf_agg.best_params_
```

```python
from sklearn.cluster import AgglomerativeClustering

agg_clustering = AgglomerativeClustering(n_clusters=clusters_agg)
df_normalized['Agglomerative_Cluster'] = agg_clustering.fit_predict(x)
silhouette_agg = silhouette_score(x, df_normalized['Agglomerative_Cluster'])
print(f"Silhouette score для Agglomerative Clustering: {silhouette_agg}")
```

```python
sns.pairplot(df_normalized, vars=["energy", "loudness"], hue='Agglomerative_Cluster', palette='magma')
plt.show()
```

```python
labels = agg_clustering.labels_
silhouette_vals = silhouette_samples(x, labels)

n_clusters = len(set(labels))
silhouette_means = [np.mean(silhouette_vals[labels == i]) for i in range(n_clusters)]

plt.figure(figsize=(8, 5))
plt.bar(range(n_clusters), silhouette_means, color='green')
plt.xlabel('Кластеры')
plt.ylabel('Средний Silhouette score')
plt.title('Средний Silhouette score по кластерам Agglomerative Clustering')
plt.show()
```

# Вывод:

# Silhouette score:
Результаты силуэтного коэффициента дают представление о том, насколько хорошо данные разделены на кластеры каждым из методов. Если коэффициент близок к 1, то объекты хорошо соответствуют своему кластеру и удалены от других кластеров, если близко к 0, то объекты находятся на границе между кластерами, если же получился отрицательный коэффициент, то это может значить, что объекты распределены неверно, так как ближе к другому кластеру.

# Общие выводы:
Silhouette score методов:

Силуэтный коэффициент для KMeans: 0.10985983514420511

Силуэтный коэффициент для Agglomerative Clustering: 0.10402780476068417

Силуэтный коэффициент для DBSCAN: 0.32543051387660943

DBSCAN лучше выделил кластеры, чем другие алгоритмы. Это может говорить о том, что данные имеют неравномерное распределение или плотностные структуры, так как DBSCAN ориентирован на плотность и способен выявлять кластеры любой формы.

Низкий силуэтный коэффициент у KMeans и Agglomerative Clustering может быть из того, что эти методы предполагают, что кластеры имеют примерно одинаковый размер или форму. И мои данные могут не соответствовать этим предположениям и, соотвественно, производительность будет хуже.

Все три метода дают положительный силуэтный коэффициент, что говорит о удовлетворительной кластеризации, потому что значения находятся ближе к нулю, чем к единице.
